# Data Visualization with Python

This notebook supports Unit 5 and uses `matplotlib`.

## 1. Bar Chart

```python
import matplotlib.pyplot as plt

subjects = ["C", "Python", "Math", "Architecture"]
marks = [78, 88, 82, 74]

plt.bar(subjects, marks)
plt.title("Subject Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()
```

## 2. Line Chart

```python
weeks = [1, 2, 3, 4, 5]
scores = [50, 58, 67, 76, 84]

plt.plot(weeks, scores, marker="o")
plt.title("Progress Over Weeks")
plt.xlabel("Week")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

## 3. Histogram

```python
marks_distribution = [35, 42, 55, 60, 60, 67, 72, 75, 78, 80, 82, 88, 91, 95]

plt.hist(marks_distribution, bins=6, edgecolor="black")
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

## 4. Scatter Plot

```python
study_hours = [1, 2, 3, 4, 5, 6, 7]
test_marks = [42, 50, 58, 65, 72, 80, 88]

plt.scatter(study_hours, test_marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

## 5. Intensive: Small Dashboard Layout

This cell creates multiple charts from one dataset. Read the titles and decide what each chart answers.

```python
students = ["Asha", "Neha", "Ravi", "Ira", "Dev", "Meera"]
marks = [82, 91, 61, 48, 76, 88]
attendance = [92, 96, 75, 68, 82, 90]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].bar(students, marks, color="steelblue")
axes[0].set_title("Marks by Student")
axes[0].set_ylabel("Marks")
axes[0].set_ylim(0, 100)

axes[1].scatter(attendance, marks, color="darkgreen")
axes[1].set_title("Attendance vs Marks")
axes[1].set_xlabel("Attendance %")
axes[1].set_ylabel("Marks")
axes[1].set_xlim(0, 100)
axes[1].set_ylim(0, 100)

plt.tight_layout()
plt.show()
```

## 6. Intensive: Distribution and Thresholds

```python
plt.hist(marks, bins=[0, 40, 60, 80, 100], edgecolor="black", color="lightgray")
plt.axvline(40, color="red", linestyle="--", label="Pass mark")
plt.title("Marks Distribution with Pass Threshold")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```
