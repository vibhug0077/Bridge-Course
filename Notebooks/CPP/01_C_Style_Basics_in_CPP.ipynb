## 2. Variables and Data Types

```c++
int age = 18;
double percentage = 86.5;
char grade = 'A';

cout << "Age: " << age << endl;
cout << "Percentage: " << percentage << endl;
cout << "Grade: " << grade << endl;
```

## 3. Operators

```c++
int a = 10;
int b = 3;

cout << "a + b = " << a + b << endl;
cout << "a / b = " << a / b << endl;
cout << "a % b = " << a % b << endl;
cout << "decimal division = " << static_cast<double>(a) / b << endl;
```

## 4. Intensive: Sizes and Type Conversion

Use this cell to connect data types with memory and arithmetic behavior.

```c++
cout << "sizeof(int) = " << sizeof(int) << " bytes" << endl;
cout << "sizeof(double) = " << sizeof(double) << " bytes" << endl;
cout << "sizeof(char) = " << sizeof(char) << " byte" << endl;

int total = 445;
int subjects = 5;
cout << "Average = " << static_cast<double>(total) / subjects << endl;
```

## 5. Intensive: Formula Program

```c++
double principal = 10000;
double rate = 7.5;
double time = 2;

double simpleInterest = (principal * rate * time) / 100;

cout << "Principal: " << principal << endl;
cout << "Simple interest: " << simpleInterest << endl;
cout << "Final amount: " << principal + simpleInterest << endl;
```
