Practice conditions, loops, lists, dictionaries, sets, and tuples.

## 1. Conditions

```python
marks = 78

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 40:
    grade = "C"
else:
    grade = "Fail"

print("Grade:", grade)
```

## 2. Loops

```python
for number in range(1, 11):
    if number % 2 == 0:
        print(number, "is even")
```

## 3. Lists

```python
marks = [82, 90, 76, 88, 95]
average = sum(marks) / len(marks)
print("Average:", average)
print("Highest:", max(marks))
```

## 4. Dictionaries

```python
student = {
    "name": "Neha",
    "branch": "CSE",
    "semester": 1,
    "marks": 88
}

for key, value in student.items():
    print(key, "=>", value)
```

## 5. Sets

```python
cities = ["Dehradun", "Delhi", "Dehradun", "Mumbai", "Delhi"]
unique_cities = set(cities)
print(unique_cities)
```

## 6. Intensive: Decision Table to Code

Validation happens before grade assignment. Change `marks` to `-5`, `39`, `75`, `90`, and `120`.

```python
marks = 92

if marks < 0 or marks > 100:
    result = "Invalid marks"
elif marks >= 90:
    result = "Grade A"
elif marks >= 75:
    result = "Grade B"
elif marks >= 40:
    result = "Pass"
else:
    result = "Fail"

print(result)
```

## 7. Intensive: Student Analytics with Nested Structures

```python
students = [
    {"name": "Asha", "branch": "CSE", "marks": [82, 91, 76]},
    {"name": "Neha", "branch": "CSE", "marks": [88, 84, 90]},
    {"name": "Ravi", "branch": "ECE", "marks": [42, 55, 61]},
    {"name": "Ira", "branch": "ME", "marks": [35, 48, 52]},
]

for student in students:
    average = sum(student["marks"]) / len(student["marks"])
    status = "Pass" if average >= 40 else "Fail"
    print(student["name"], student["branch"], f"average={average:.2f}", status)
```

## 8. Intensive: Counting with Dictionaries

```python
sentence = "data science uses data"
counts = {}

for char in sentence.replace(" ", ""):
    counts[char] = counts.get(char, 0) + 1

print(counts)
```
